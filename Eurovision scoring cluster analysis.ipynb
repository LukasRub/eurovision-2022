#### Reading data

```python
import numpy as np
import pandas as pd
from IPython.display import display

scores = pd.read_excel("./scores.ods", index_col=0)
display(scores)
```

### Hierarchical clustering

```python
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, dendrogram

pairwise_dists = pdist(scores.T, metric="euclidean")
col_clusters = linkage(pairwise_dists, method="complete")

plt.figure(figsize=(10,6))
col_dendrogram = dendrogram(col_clusters, labels=scores.columns)
plt.tight_layout()
plt.ylabel("Euclidean distance")
plt.show()
```

### Dimensionality reduction

```python
# Center data
X = scores.T.values
X_centered = X - np.mean(X, axis=0)

# Perform SVD
U, S, Vt = np.linalg.svd(X_centered, full_matrices=False)

# Calculate explained variance ratio
explained_variance = (S**2) / (len(scores) - 1)
total_variance = explained_variance.sum()
explained_variance_ratio = explained_variance / total_variance

# Set up the plot
fig, ax = plt.subplots(figsize=(10, 12))
x = np.arange(1, len(explained_variance_ratio)+1)
y = np.arange(0, 1+0.1, 0.1)
plt.xticks(x)
plt.yticks(y)

# Plot variance explained by each principal component 
bar_plot = plt.bar(x, explained_variance_ratio, 
                   label="Explained variance ratio by each individual component")
bar_labels = [format(evr, ".4f") for evr in explained_variance_ratio]
ax.bar_label(bar_plot, bar_labels, padding=3)

# Plot cumulative sum of explained variance by each included component
plt.plot(x, explained_variance_ratio.cumsum(), "d-", color="orange",
        label="Cumulative explained variance ratio if component is included")
for x, y in zip(x, explained_variance_ratio.cumsum()):
    if x != 1:
        ax.annotate(format(y, ".4f"), (x, y), 
                    ha='center', va='center',
                    xytext=(0, 8),
                    textcoords='offset points')
    

plt.xlabel("Principal component")
plt.ylabel("Explained variance ratio")
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X_reduced = pca.fit_transform(scores.T)

pairwise_dists = pdist(X_reduced, metric="euclidean")
col_clusters = linkage(pairwise_dists, method="complete")

plt.figure(figsize=(10,6))
col_dendrogram = dendrogram(col_clusters, labels=scores.columns)
plt.tight_layout()
plt.ylabel("Euclidean distance")
plt.show()
```
